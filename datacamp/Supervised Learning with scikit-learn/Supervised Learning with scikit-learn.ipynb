# 1. Classification  

1- Build a model  
2- Model learns from the labeled data we pass to it  
3- pass the unlabeled data to the model as input  
4- Model predicts the labels of the unseen data  

**k-Nearest Neighbors**  

- Predict the label of a data point by
  - Looking at the **k** closest labeled data points
  - Taking a majority vote










```python
import numpy as np
import pandas as pd
```

```python
# read the data
churn_df = pd.read_csv("telecom_churn_clean.csv")
churn_df
```

```python
churn_df.describe()
```

```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier
# X is a 2-d array of our features
X = churn_df[["total_day_charge", "total_eve_charge"]] # to subset two columns
X = churn_df[["total_day_charge", "total_eve_charge"]].values # # X is a 2-d array of our features
y = churn_df["churn"].values 
# check and confirm the size
print(X.shape, y.shape)
```

```python
# next step, assign the variable or estantuate the mode
knn = KNeighborsClassifier(n_neighbors= 15)
# fit the classifier 
knn.fit(X,y)
```

```python
# predicting on unlabeled data

X_new = np.array([[56.8, 17.5],
                  [24.4, 24.1],
                  [50.1, 10.9]])
```

```python
# predice the new or unseen data

predictions = knn.predict(X_new)
predictions
```

```python
print ('predictions: {}'.format(predictions))
```

```python
# all steps in one chunck

# Import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier 

# Create arrays for the features and the target variable
y = churn_df["churn"].values
X = churn_df[["account_length", "customer_service_calls"]].values

# Create a KNN classifier with 6 neighbors
knn = KNeighborsClassifier(n_neighbors= 6)

# Fit the classifier to the data
knn.fit(X, y)

# Predict the labels for the X_new
y_pred = knn.predict(X_new)

# Print the predictions for X_new
print("Predictions: {}".format(y_pred)) 
```

## Assessing Accuracy of the Model

```python
# Train/tes split

from sklearn.model_selection import train_test_split

# 30% as as test set
# stratify = y to ensure our split to account for the actual proportion for our data.

X_train, X_test, y_train, y_test = train_test_split (X,y, test_size = 0.3, random_state = 21, stratify=y)

# This will return 4 arrays (X_train, X_test, y_train, y_test)

# Next, we will fit the knn model 

knn = KNeighborsClassifier(n_neighbors= 6)
knn.fit(X_train, y_train)

# To assess the model's accuracy, we use the method knn.score(X_Test, y_test)

print(knn.score(X_test, y_test))
```

```python
# the number of k_neighnors is a balance between over fitting and underfitting. More k neighbors means a less complex model but might lead to overfitting. On the other 
# hand, a smaller number of neighbors means a more complex model and leads to overfitting.

```

```python
from sklearn import neighbors
# To ptimize the the number of neibors, we do a loop over the potential values for the number of neeigbors

# crate two empty dictionaries.

train_accuracies = {}
test_accuracies = {}

# for loop the potential values
## create the number of loops
neighbors = np.arange(1,26) # 1D array from from 1 to 25
neighbors = np.arange(1,26)

for i in neighbors:
  knn = KNeighborsClassifier(n_neighbors= i)
  knn.fit(X_train, y_train)
  train_accuracies[i] = knn.score(X_train,y_train)
  test_accuracies[i] = knn.score(X_train,y_train)
```

```python
train_accuracies
train_accuracies.values()
```

```python
# plot the results
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (8,6))
plt.title("KNN: Varying Number of Neighbors")
plt.plot(neighbors, neighbors, label="Training Accuracy")
#plt.plot(neighbors, test_accuracies.values(), label="Testing Accuracy")
plt.legend()
plt.xlabel("Number of Neighbors")
plt.ylabel("Accuracy")
plt.show()
```
